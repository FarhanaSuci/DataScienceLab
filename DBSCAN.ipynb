# Select Relevant feature for Clustering

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

# Apply DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
```

```python
epsilon = 0.5
```

```python
minimum_sample = 5
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
```

```python
labels = dbscan.fit_predict(X_scaled)
```

```python
labels
```

##  Add Cluster label to DataFrame

```python
df['Cluster'] = labels
```

```python
df.tail(15)
```

# Identify Crore, Border and Noise Points

```python
import numpy as np
```

```python
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True  # Core points
```

# Count Points

```python
num_noise = np.sum(labels == -1)  
```

```python
num_core = np.sum(core_samples_mask)  

```

```python
num_border = len(labels) - (num_core + num_noise) 
```

## Crore Points

```python
print(f"Number of Core Points: {num_core}")
```

## Border Points

```python
print(f"Number of Border Points: {num_border}")
```

## Noise Points

```python
print(f"Number of Noise Points: {num_noise}")
```

# Plot the  core, border, and noise points

```python
plt.figure(figsize=(10, 6))
for label in set(labels):
    if label == -1:
        color = 'black'  # Noise points
        label_name = "Noise"
    else:
        color = np.random.rand(3,)  # Random color for each cluster
        label_name = f"Cluster {label}"
    
    mask = (labels == label)
    plt.scatter(X_scaled[mask & core_samples_mask, 0], X_scaled[mask & core_samples_mask, 1], 
                c=[color], label=label_name + " (Core)", edgecolors='k', marker='o', s=100)
    plt.scatter(X_scaled[mask & ~core_samples_mask, 0], X_scaled[mask & ~core_samples_mask, 1], 
                c=[color], edgecolors='k', marker='x', s=100, label=label_name + " (Border)")

plt.scatter(X_scaled[labels == -1, 0], X_scaled[labels == -1, 1], 
            c='black', label="Noise", edgecolors='k', marker='s', s=100)

plt.title("Core, Border, and Noise Points")
plt.xlabel("Annual Income ( Standardization )")
plt.ylabel("Spending Score ( Standardization )")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Annual Income (k$)'], y=df['Spending Score (1-100)'], hue=df['Cluster'], palette='Set1', s=100, edgecolor='k')

plt.title("DBSCAN Clustering")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend(title="Cluster", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
```

```python


neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X_scaled)
distances, indices = neighbors_fit.kneighbors(X_scaled)

distances = np.sort(distances[:, 4])  # 4th neighbor distance
plt.plot(distances)
plt.xlabel("Points sorted by distance")
plt.ylabel("4th Nearest Neighbor Distance")
plt.title("Elbow Method for Choosing epsilon")
plt.show()
```
